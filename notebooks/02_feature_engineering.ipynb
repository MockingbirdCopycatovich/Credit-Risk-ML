```python
import sys
from pathlib import Path

# PROJECT_DIR points to the root of your project
PROJECT_DIR = Path().resolve().parent
sys.path.append(str(PROJECT_DIR))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.data_loader import load_application_train
from src.preprocessing import clean_data

# Enable inline plots in Jupyter
%matplotlib inline
sns.set(style="whitegrid")
```

```python
# Load application_train.csv using data_loader
# clean_data adds AGE, missing value flags, and new feature ratios
df = load_application_train(PROJECT_DIR / "data" / "raw")
df = clean_data(df)
```

```python
# CREDIT_INCOME_RATIO: Credit amount divided by income
# Shows the distribution of credit relative to income for target=0 vs target=1
plt.figure(figsize=(10, 6))
sns.boxplot(x="TARGET", y="CREDIT_INCOME_RATIO", data=df)
plt.yscale("log")
plt.title("CREDIT_INCOME_RATIO by TARGET")
plt.xlabel("TARGET (0 = no default, 1 = default)")
plt.ylabel("Credit / Income (log scale)")
plt.show()
```

```python
# ANNUITY_CREDIT_RATIO: Monthly annuity divided by total credit
# Shows if higher monthly payments relative to credit amount correlate with default
plt.figure(figsize=(10, 6))
sns.boxplot(x="TARGET", y="ANNUITY_CREDIT_RATIO", data=df)
plt.title("ANNUITY_CREDIT_RATIO by TARGET")
plt.xlabel("TARGET (0 = no default, 1 = default)")
plt.ylabel("Annuity / Credit")
plt.show()
```

```python
# EMPLOYED_AGE_RATIO: Days employed divided by age
# Shows employment length relative to age for each target class
plt.figure(figsize=(10, 6))
sns.boxplot(x="TARGET", y="EMPLOYED_AGE_RATIO", data=df)
plt.title("EMPLOYED_AGE_RATIO by TARGET")
plt.xlabel("TARGET (0 = no default, 1 = default)")
plt.ylabel("Employed / Age")
plt.show()
```
